#### Manually calculate standard deviation

$$\sigma = \sqrt{\frac{1}{n}\sum_{i=1}^n(x_i - \mu)^2}$$

- Write a function to calculate standard deviation using the formula above

````python
# Calculate standard deviation given a data input
def calc_std(data):
    mean = data.mean()
    sum_of_squared_difference = 0
    for datum in data:
        squared_difference = (datum - mean) ** 2
        sum_of_squared_difference += squared_difference
    return (sum_of_squared_difference / len(data)) ** 0.5
````

- Use a **dictionary comprehension** to apply your standard deviation function to each numeric column in the dataframe.  **No loops**  
- Assign the output to variable `sd` as a dictionary where: 
    - Each column name is now a key 
    - That standard deviation of the column is the value 
     
*Example Output :* `{'ACT_Math': 120, 'ACT_Reading': 120, ...}`

````python
# Display standard deviations calculated using the formula above
sd = {title: calc_std(final[title]) for title in final.columns if title != 'state'}

# Import pprint to display the standard deviations
import pprint
pprint.pprint(sd)
````

````python
# Display standard deviations calculated using the Pandas describe method
pandas_sd = final.describe().transpose()['std']
print(pandas_sd)
````

````python
# Display standard deviations calculated using the numpy.std method
numpy_sd = {title: np.std(final[title]) for title in final.columns if title != 'state'}
pprint.pprint(numpy_sd)
````

- The manually calculated standard deviation and numpy's std method are equivalent. They use sample size n in the denominator.

#### Define a custom function to subplot histograms

We have data for two tests for two years. We only have composite (and not subtest scores) for the 2018 ACT. We should write a function that will take the names of 2+ columns and subplot histograms. While you can use pandas plotting or Seaborn here, matplotlib gives you greater control over all aspects of your plots.

[Helpful Link for Plotting Multiple Figures](https://matplotlib.org/users/pyplot_tutorial.html#working-with-multiple-figures-and-axes)

Here's some starter code:

````python
def subplot_histograms(dataframe, list_of_columns, list_of_titles, list_of_xlabels):
    nrows = int(np.ceil(len(list_of_columns)/2))                                      # Makes sure you have enough rows
    fig, ax = plt.subplots(nrows=nrows, ncols=2, figsize =(10,10),sharey=True)    # You'll want to specify your figsize
    ax = ax.ravel()                                    # Ravel turns a matrix into a vector, which is easier to iterate
    for i, column in enumerate(list_of_columns):                    # Gives us an index value to get into all our lists
        ax[i].hist(dataframe[column])                                                  # feel free to add more settings
        ax[i].title.set_text(f'{list_of_titles[i]}')                    # Set titles, labels, etc here for each subplot
        ax[i].set_xlabel(f'{list_of_xlabels[i]}')
````

````python
# Create a function to easily plot a boxplot given a dataframe and list of column titles
def subplot_boxplots(dataframe, list_of_columns):
    nrows = int(np.ceil(len(list_of_columns)/2))                                
    fig, ax = plt.subplots(nrows=nrows, ncols=2, figsize =(10,5),sharey=True)
    ax = ax.ravel()                                    
    for i, column in enumerate(list_of_columns):
        ax[i].boxplot(dataframe[column])
        ax[i].title.set_text(f'{list_of_columns[i]}')
````

#### Statistical Evaluation of Distributions 

**If you feel it's appropriate**, using methods we discussed in class, run hypothesis tests to compare variables of interest in our dataset. 

````python
# Do states that have high SAT participation rates perform better than states which do not?
# Perform hypothesis testing using the 2018 data

# Create dataframe for states that have high (>= 80%) participation rates
high_participation = final.loc[final['sat_participation_2018'] >= 80,
                         ['state', 'sat_total_2018', 'sat_participation_2018']]

# Create series for SAT total scores of high participation states
high_participation_scores = high_participation.loc[:,'sat_total_2018']
````

````python
# Create dataframe for all other states
non_high_participation = final.loc[(final['state'].isin (set(final['state']) - set(high_participation['state']))),
                         ['state', 'sat_total_2018', 'sat_participation_2018']]

# Create series for SAT total scores other states
non_high_participation_scores = non_high_participation.loc[:,'sat_total_2018']
````

````python
# Null hypothesis:
# There is no difference in mean SAT total scores between states with high SAT participation rates and other states

# Conduct t-test
stats.ttest_ind(high_participation_scores,non_high_participation_scores, equal_var=False)
````

````python
print(high_participation_scores.mean())
print(non_high_participation_scores.mean())
````

Based on pvalue = 4.85e-07 < 0.05, we reject the null hypothesis that there is no difference in mean SAT total scores between states with high SAT participation rates and other states. 

There is very low probability that we observe mean total SAT scores of 1033 and 1148 (for states with high participation rates vs other states respectively) assuming our null hypothesis is true. We thus have to reject the null hypothesis.
